# Create binary mask for monthly SOCAT data

```python
from mpl_toolkits.basemap import Basemap # for plotting 2D data on maps
from mpl_toolkits import basemap
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt
import seaborn as sns
import cmocean
import os 

from matplotlib.patches import Polygon
from matplotlib.collections import PatchCollection
from matplotlib.patches import PathPatch
from mpl_toolkits.axes_grid1 import make_axes_locatable
%matplotlib inline
```

```python
dataDir = '/Users/gloege/Documents/Projects/SOCAT-neural-network/data/data_netcdf/'
ds = xr.open_dataset(dataDir + 'SOCATv5_mask.nc')

### Load variables and put negative shit at end
### This is necessary to lon go 0-360
lon = np.arange(0.5, 360.5)
lat = ds['ylat']
time = ds['tmnth']
mask = ds['socat_mask'].roll(xlon=-180)

obs = np.squeeze(np.sum(mask, axis=0).values)
#obs[(obs==0)]=np.NaN
```

```python
### Setup figure
fig = plt.figure(figsize=(8.5,11))
ax = fig.add_subplot(111)

### create meshgrid
lon2, lat2 = np.meshgrid(lon, lat)

### define the colormap
cmap = cmocean.cm.balance

### Generate a basemap
m = Basemap(resolution='l', projection='cyl', ax=ax)

### convert data meshgrid to basemap coordinates
x, y = m(lon2, lat2)

### Draw coastlines
m.drawcoastlines

### Uncomment this to fill in continents with a color
#m.fillcontinents(color='black',lake_color='aqua')

### Plot the data on the basemap
sub = m.pcolor(x, y, np.ma.masked_array(obs, mask = (obs==0)), 
               cmap = cmap, 
               #norm=norm,
               vmin = 0, 
               vmax = 100,
               latlon=True)

#-------------------------------------------------------
# Colorbar: right side, width=7% of ax, fixed at 0.05 in 
# Notes : sub is the plotted data (comes from subplot)
#-------------------------------------------------------
divider = make_axes_locatable(ax)
cax = divider.append_axes("right", size="7%", pad=0.05)
col = fig.colorbar(sub, cax=cax)
col.set_label(r'Number of observations', fontsize=14)
#col.set_ticks([np.arange(0,6,0.5)])
col.ax.tick_params(labelsize=12) 

### show the figure
plt.show()
```

## Create new NetCDF file

```python
### Create netcdf file for socat mask
ds = xr.Dataset(
    {
    'socat_mask': (['tmnth','ylat', 'xlon'],  mask[144:, :, :]),
    },

    coords={
    'ylat': (lat),
    'xlon': (lon),
    'tmnth': time[144:]
    }

    )
```

```python
print('write output')
ds.to_netcdf('SOCATv5_mask_1982-2017.nc')
```
