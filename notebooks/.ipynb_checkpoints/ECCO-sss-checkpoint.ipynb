# ECCO sea surface salinity

```python
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt
import matplotlib as mpl
from mpl_toolkits.basemap import Basemap  # for plotting 2D data on maps
from mpl_toolkits import basemap
from matplotlib.patches import Polygon
from matplotlib.collections import PatchCollection
from matplotlib.patches import PathPatch
from mpl_toolkits.axes_grid1 import make_axes_locatable
import cmocean
import seaborn as sns
import os
%matplotlib inline
```

```python
dataDir = '/Users/gloege/Documents/Projects/SOCAT-neural-network/data/data_netcdf/'
fileNames = os.listdir(dataDir)
print(fileNames)
```

```python
ds = xr.open_dataset(dataDir + 'SSS_monthly_1deg_1992-2015.nc')
```

```python
lon=ds['xlon'].values
lat=ds['ylat'].values
sss=ds['SALT'].values
```

```python
sss_mn=np.nanmean(sss,axis=0)
lon
```

```python

### Setup figure
fig = plt.figure(figsize=(8.5,11))
ax = fig.add_subplot(111)

### create meshgrid
lon2, lat2 = np.meshgrid(lon, lat)

### define the colormap
cmap = cmocean.cm.balance

### Generate a basemap
m = Basemap(resolution='l', projection='cyl', ax=ax)

### convert data meshgrid to basemap coordinates
x, y = m(lon2, lat2)

### Draw coastlines
m.drawcoastlines

### Uncomment this to fill in continents with a color
#m.fillcontinents(color='black',lake_color='aqua')

### Plot the data on the basemap
sub = m.pcolor(x, y, np.ma.masked_array(sss_mn, mask = (sss_mn==0)), 
               cmap = cmap, 
               #norm=norm,
               vmin = 33, 
               vmax = 38)


#-------------------------------------------------------
# Colorbar: right side, width=7% of ax, fixed at 0.05 in 
# Notes : sub is the plotted data (comes from subplot)
#-------------------------------------------------------
divider = make_axes_locatable(ax)
cax = divider.append_axes("right", size="7%", pad=0.05)
col = fig.colorbar(sub, cax=cax)
col.set_label(r'sss [psu]', fontsize=14)
#col.set_ticks([np.arange(0,6,0.5)])
col.ax.tick_params(labelsize=12) 

### show the figure
plt.show()
```
