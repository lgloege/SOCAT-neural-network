```python
import tensorflow as tf
import numpy as np 
from sklearn.preprocessing import scale
from matplotlib import pyplot as plt
```

```python
### Input features 
x = tf.placeholder(shape=[None, 5], dtype=tf.float32)

### Hidden Layers
nn = tf.layers.dense(x, 3, activation=tf.nn.relu)
nn = tf.layers.dense(nn, 5, activation=tf.nn.relu)
nn = tf.layers.dense(nn, 5, activation=tf.nn.relu)

### Output layer
nn = tf.layers.dense(nn, 1, activation=None)
```

```python
### Cost, Optimizer, and 
cost = tf.reduce_mean((nn - x)**2)
optimizer =tf.train.AdamOptimizer(1e-4).minimize(cost)
init = tf.global_variables_initializer()
#init = tf.initialize_all_variables()
```

```python
tf.summary.scalar("cost", cost)
merged_summary_op = tf.summary.merge_all()
```

```python
with tf.Session() as sess:
    sess.run(init)
    #uniq_id = "/tmp/tensorboard-layers-api/" + uuid.uuid1().__str__()[:6]
    #summary_writer = tf.summary.FileWriter(uniq_id, graph=tf.get_default_graph())
    x_vals = np.random.normal(0, 1, (10000, 5))
    for step in range(10000):
        _, val, summary = sess.run([optimizer, cost, merged_summary_op],
                                   feed_dict={x: x_vals})
        if step % 5 == 0:
            print("step: {}, value: {}".format(step, val))
            #summary_writer.add_summary(summary, step)
```
