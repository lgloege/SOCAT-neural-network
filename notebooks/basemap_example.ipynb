```python
lon = ds['lon'].values
lat = ds['lat'].values
time = ds['time'].values
chl = ds['Chl'].values
```

```python
chl.shape
```

```python
### take the mean over time
chl_mn = np.nanmean(chl, axis=0)
```

# 3. Generate a figure
I use Basemap to make maps, here are some basemap resources

https://jakevdp.github.io/PythonDataScienceHandbook/04.13-geographic-data-with-basemap.html

https://matplotlib.org/basemap/users/examples.html

```python
### Setup figure
fig = plt.figure(figsize=(8.5,11))
ax = fig.add_subplot(111)

### create meshgrid
lon2, lat2 = np.meshgrid(lon, lat)

### define the colormap
cmap = cmocean.cm.balance

### Generate a basemap
m = Basemap(resolution='l', projection='cyl', ax=ax)

### convert data meshgrid to basemap coordinates
x, y = m(lon2, lat2)

### Draw coastlines
m.drawcoastlines

### Uncomment this to fill in continents with a color
#m.fillcontinents(color='black',lake_color='aqua')

### Plot the data on the basemap
sub = m.pcolor(x, y, np.ma.masked_array(chl_mn, mask = (chl_mn==0)), 
               cmap = cmap, 
               #norm=norm,
               vmin = 0, 
               vmax = 1)


#-------------------------------------------------------
# Colorbar: right side, width=7% of ax, fixed at 0.05 in 
# Notes : sub is the plotted data (comes from subplot)
#-------------------------------------------------------
divider = make_axes_locatable(ax)
cax = divider.append_axes("right", size="7%", pad=0.05)
col = fig.colorbar(sub, cax=cax)
col.set_label(r'Chl [mg$m^{-3}$]', fontsize=14)
#col.set_ticks([np.arange(0,6,0.5)])
col.ax.tick_params(labelsize=12) 

### show the figure
plt.show()
```
